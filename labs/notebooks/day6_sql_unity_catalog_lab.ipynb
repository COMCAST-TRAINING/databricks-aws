# Day 6 Lab: SQL Warehouse & Unity Catalog

## Objectives
- Query Gold layer data using SQL
- Understand SQL Warehouse usage
- Implement governance using Unity Catalog


```
%sql
SELECT * FROM gold_customer_spend ORDER BY total_spent DESC
```

```
%sql
CREATE CATALOG IF NOT EXISTS demo_catalog;
CREATE SCHEMA IF NOT EXISTS demo_catalog.sales;
```

```
%sql
CREATE TABLE demo_catalog.sales.gold_customer_spend AS SELECT * FROM gold_customer_spend;
```

```
%sql
SELECT * FROM demo_catalog.sales.gold_customer_spend;
```

```
%sql
GRANT SELECT ON TABLE demo_catalog.sales.gold_customer_spend TO `analyst_role`;
```

```
%sql
SHOW GRANTS ON TABLE demo_catalog.sales.gold_customer_spend;
```
